# outliers_analysis

```python
for i in df_1.columns:
    sns.boxplot(x=i,data=df_1)
    plt.show()
    # it is visible in the plots the we certainly have outliers in our data
    # we process them by using flooring and capping method
    
```

```python
#handling outliers by flooring and capping
for column in df_1.columns:
    q1 = df_1[column].quantile(0.25)
    q3 = df_1[column].quantile(0.75)
    iqr = q3 - q1
    ll = q1 - (1.5*iqr)
    ul = q3 + (1.5*iqr)
    #print(column,ll,ul)
    for index in df_1[column].index:
        if df_1.loc[index,column]>ul:
            df_1.loc[index,column]=ul
        if df_1.loc[index,column]<ll:
            df_1.loc[index,column]=ll
```

# scailing

```python
x= df_1.drop(columns='selling_price') # independent variable
y= df_1['selling_price'] # dependent varaible 
# we scaled only target variable , we do not scale target variable
```

```python
from sklearn.preprocessing import StandardScaler #standardization 
#!pip install sklearn
ss=StandardScaler()
x_scaled = pd.DataFrame(ss.fit_transform(x),index=x.index,columns=x.columns)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=10)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
x_test.head()
```

```python
y_test.head()
```

# linear reagression model

```python
x_train_const = sm.add_constant(x_train)# Adding constant will give me the intercept of the model
# we add constant because of the intercept 
x_train_const.head()
```

```python
lr_ols = sm.OLS(y_train,x_train_const).fit()
lr_ols.summary()
```

```python
# if km_driven increasess by 1 unit then selling price of the car deacreses by the value of 3.694
# const are the intercept rest of them are the slope 
# if torque increases  by i unit then selling price of the car increases by the value of 1896.1609
```

```python
# interpreting the p-value 
# All the column except torque are related to the selling price 
```

# performance matrix

```python
x_test_const = sm.add_constant(x_test)
y_test_pred = lr_ols.predict(x_test_const)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_percentage_error
mse = mean_squared_error(y_test,y_test_pred)
print('mse', mse)
print('rmse',np.sqrt(mse))
mape = mean_absolute_percentage_error(y_test,y_test_pred)
print('mape', mape*100)
```

# train and test score 

```python
y_test_pred = lr_ols.predict(x_test_const) # giving the new data
y_train_pred = lr_ols.predict(x_train_const) #giving the same data that it is trained on 
```

```python
print('Train RMSE',np.sqrt(mean_squared_error(y_train,y_train_pred)))
print('Test RMSE',np.sqrt(mean_squared_error(y_test,y_test_pred)))
```

```python
print('Train mapE',np.sqrt(mean_absolute_percentage_error(y_train,y_train_pred)))
print('Test mapE',np.sqrt(mean_absolute_percentage_error(y_test,y_test_pred)))
```

```python
from sklearn.metrics import r2_score
print('Train r2',r2_score(y_train,y_train_pred))
print('Test r2',r2_score(y_test,y_test_pred))
```

```python
# there is no sign of underfitting and overfitting of the data judging by the train test score
```
